```python
def image_processor(image_path, target_size):
    """Preprocess images for CNN model"""
    absolute_image_path = os.path.abspath(image_path)
    image = cv2.imread(absolute_image_path)
    image = cv2.resize(image, (target_size[1], target_size[0]))
    if target_size[2] == 1:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Normalize image (mean normalization)
    image_array = image.astype('float32') / 255.0
    
    return image_array    
```

## Build Generator and Discriminator Models

```python
# # Generator model
def build_generator(latent_dim, num_classes, img_shape, generator_lr=0.0001):
    noise = Input(shape=(latent_dim,))
    label = Input(shape=(1,))
    
    # Label embedding
    label_embedding = Flatten()(Embedding(num_classes, latent_dim)(label))

    # Concatenate noise and label embeddings
    model_input = concatenate([noise, label_embedding])
    
    # Project combined input to higher dimension
    model = Dense(128 * 7 * 7)(model_input)
    model = Reshape((7, 7, 128))(model)

    # Upsampling blocks
    model = Conv2DTranspose(32, kernel_size=4, strides=2, padding='same')(model) #14
    model = BatchNormalization()(model)
    model = LeakyReLU(alpha=0.2)(model)

    model = Conv2DTranspose(64, kernel_size=4, strides=2, padding='same')(model) #28
    model = BatchNormalization()(model)
    model = LeakyReLU(alpha=0.2)(model)

    # Additional Upsampling layers to reach 224x224
    model = Conv2DTranspose(128, kernel_size=4, strides=2, padding='same')(model) #56
    model = BatchNormalization()(model)
    model = LeakyReLU(alpha=0.2)(model)
    
    model = Conv2DTranspose(128, kernel_size=4, strides=1, padding='same')(model) # 56
    model = BatchNormalization()(model)
    model = LeakyReLU(alpha=0.2)(model)

    model = Conv2DTranspose(256, kernel_size=4, strides=2, padding='same')(model) # 112
    model = BatchNormalization()(model)
    model = LeakyReLU(alpha=0.2)(model)
    
    # Upsample to reach the target resolution (224x224)
    model = Conv2DTranspose(img_shape[2], kernel_size=4, strides=2, padding='same', activation='tanh')(model)
    # model = Lambda(lambda x: tf.image.resize(x, (224, 224)))(model)

    generator = Model([noise, label], model)

    generator_optimizer = Adam(learning_rate=generator_lr)
    # Compile the generator model with the optimizer
    generator.compile(loss='binary_crossentropy', optimizer=generator_optimizer)

    return generator
```

```python
latent_dim = 128
num_classes = len(full_mass['labels'].unique())
img_size = (224, 224)
channels = target_size[2]
img_shape = (img_size[0], img_size[1], channels)
class_count = 2 # to define number of classes in dense layer
```

```python
# Build the generator
generator = build_generator(latent_dim, class_count, 
                            img_shape, generator_lr=0.0002)

# The generator takes noise and the target label as input and generates an image
z = Input(shape=(latent_dim,))
label = Input(shape=(1,))
img = generator([z, label])

#Check the image shape:
print("Image output: \n", img)
generator.summary()
```

```python
from keras.optimizers import Adam

# Discriminator model
def build_discriminator(img_shape, num_classes):
    image = Input(shape=img_shape)
    label = Input(shape=(1,))

    label_embedding = Flatten()(Embedding(num_classes, np.prod(img_shape))(label))
    label_embedding = Reshape(img_shape)(label_embedding)

    model_input = Concatenate()([image, label_embedding])

    model = Conv2D(16, kernel_size=3, strides=2, padding="same")(model_input)
    # model = BatchNormalization()(model)
    # model = LeakyReLU(alpha=0.2)(model)
    
    # model = Conv2D(16, kernel_size=3, strides=1, padding="same")(model)
    # # model = BatchNormalization()(model)
    model = LeakyReLU(alpha=0.2)(model)


    model = Flatten()(model)
    model = Dropout(0.5)(model)
    
    validity = Dense(1, activation='sigmoid')(model)

    discriminator = Model([image, label], validity)

    return discriminator
```

```python
# Build and compile the discriminator
discriminator_optimizer = Adam(learning_rate=0.0001, beta_1=0.5, clipvalue=0.5)  # Adjust clipvalue as needed
discriminator = build_discriminator(img_shape, class_count)
discriminator.compile(loss='binary_crossentropy',
                      optimizer=discriminator_optimizer,
                      metrics=['accuracy'])
```

```python
discriminator.summary()
```

```python
# For the combined model, only train the generator
discriminator.trainable = False

# The discriminator takes generated image and the target label as input and determines validity
valid = discriminator([img, label])

# The combined model (stacked generator and discriminator)
# Trains the generator to fool the discriminator
combined_model = Model([z, label], valid)
combined_model.compile(loss='binary_crossentropy', optimizer=Adam(0.000075, 0.5))
combined_model.summary()

batch_size = 32
half_batch = batch_size // 2
```

```python
def training_cgan(epochs, Xtrain, ytrain):
    for epoch in range(epochs):
        # Train discriminator
        idx = np.random.randint(0, Xtrain.shape[0], half_batch)
        # imgs, labels = Xtrain[idx], ytrain[idx]

        noise = np.random.normal(0, 1, (half_batch, latent_dim))
        labels_for_gen = ytrain[idx][:, 0].reshape((-1, 1))  # Considering the first column of y_train

        gen_imgs = generator.predict([noise, labels_for_gen])

        # Create labels for real and fake samples
        labels_real = np.ones((half_batch, 1))
        labels_fake = np.zeros((half_batch, 1))

        d_loss_real = discriminator.train_on_batch([Xtrain[idx], labels_for_gen], labels_real)

        d_loss_fake = discriminator.train_on_batch([gen_imgs, labels_for_gen], labels_fake)

        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

        # Train generator
        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        sampled_labels = np.random.randint(0, class_count, batch_size).reshape(-1, 1)

        g_loss = combined_model.train_on_batch([noise, sampled_labels], np.ones((batch_size, 1)))

        # Print progress
        if epoch % 10 == 0:
            print(f"Epoch {epoch}: [D loss: {d_loss[0]}, acc.: {100 * d_loss[1]}] [G loss: {g_loss}]")
```

```python
# Generating new images
def generate_images(n_rows, n_cols, generator_model):
    noise = np.random.normal(0, 1, (n_rows * n_cols, latent_dim))
    #labels = np.tile(np.arange(0, 10).reshape(-1, 1), n_rows).reshape(-1, 1)
    labels = np.random.randint(0, num_classes, n_rows * n_cols).reshape(-1, 1)

    generated_images = generator_model.predict([noise, labels])
    generated_images = 0.5 * generated_images + 0.5  # Rescale images 0 - 1
    return generated_images
```

```python
def generate_and_save_images_with_labels(n_images, generator_model, output_folder, class_label, start_index=0):
    os.makedirs(output_folder, exist_ok=True)

    generated_images = []
    generated_labels = []

    for i in range(n_images):
        noise = np.random.normal(0, 1, (1, latent_dim))
        label = np.full((1, 1), class_label)  # Use specified class label

        generated_image = generator_model.predict([noise, label])
        generated_image = 0.5 * generated_image + 0.5  # Rescale images 0 - 1

        # Convert to PIL image
        generated_image_pil = Image.fromarray((generated_image[0] * 255).astype(np.uint8))

        # Save the image in JPEG format with an incremented index
        image_path = os.path.join(output_folder, f'{start_index}_generated_image_{i}.png')
        generated_image_pil.save(image_path, 'PNG')

        
        label_class = np.array([1.0, 0.0]) if class_label == 0 else np.array([0.0, 1.0])
        generated_images.append(generated_image[0])
        generated_labels.append(label_class)

    # Convert to numpy arrays
    generated_images = np.array(generated_images)
    generated_labels = np.array(generated_labels)

    return generated_images, generated_labels
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# create labels for confusion matrix
cm_labels = ['MALIGNANT', 'BENIGN']

# obtain predictions
y_pred_test = cnn_model.predict(X_test)
y_pred_train = cnn_model.predict(X_new)

# convert predicted probabilities to class predictions
y_pred_classes_test = np.argmax(y_pred_test, axis=1)
y_pred_classes_train = np.argmax(y_pred_train, axis=1)

# Assuming y_test and y_val are in binary format (0 or 1)
y_true_classes_test = np.argmax(y_test, axis=1)
y_true_classes_train = np.argmax(y_new, axis=1)

# generate classification reports for test and val sets
test_report = classification_report(y_true_classes_test, y_pred_classes_test, target_names=cm_labels)
train_report = classification_report(y_true_classes_train, y_pred_classes_train, target_names=cm_labels)

# generate confusion matrices for test and validation sets
test_cm = confusion_matrix(y_true_classes_test, y_pred_classes_test)
train_cm = confusion_matrix(y_true_classes_train, y_pred_classes_train)

# create function to print confusion matrix
def plot_confusion_matrix(cm, labels, title):
    """plots a normalized confusion matrix as a heatmap."""
    # Calculate row sums
    row_sums = cm.sum(axis=1, keepdims=True)
    # Normalize confusion matrix
    normalized_cm = cm / row_sums

    plt.figure(figsize=(8, 6))
    sns.heatmap(normalized_cm, annot=True, fmt='.2%', cmap='Blues', cbar=False,
                xticklabels=labels, yticklabels=labels)
    plt.title(title, fontsize=14)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

# print Train and Test reports and matrices
print(f"Train Set Classification report:\n {train_report}\n")
plot_confusion_matrix(train_cm, cm_labels, 'Train Set Confusion Matrix')
```

```python
print(f"Test Set Classification report:\n {test_report}\n")
plot_confusion_matrix(test_cm, cm_labels, 'Test Set Confusion Matrix')
```
